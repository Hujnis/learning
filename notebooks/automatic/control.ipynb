Inspirace z [videa](https://www.youtube.com/watch?v=2lQbGQ_cQ3w)


```
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import signal
%matplotlib inline
```

## Definice lineárního systému

```
Kp = 3.0
taup = 2.0
```

```
num = [Kp]
den = [taup, 1]
sys = signal.TransferFunction(num, den)
```

## Odezva systému na jednotkový skok

```
t, y = signal.step(sys)
plt.plot(t, y)
plt.grid(which='both')
```

## Frekvenční charakteristika systému

```
w, mag, phase = signal.bode(sys)
plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.grid(which='both')

plt.figure()
plt.semilogx(w, phase)  # Bode phase plot
plt.grid(which='both')
plt.show()
```

## Odezva systému na jednotkový impuls

```
t, y = signal.impulse(sys)
plt.plot(t, y)
plt.grid(which='both')
```
