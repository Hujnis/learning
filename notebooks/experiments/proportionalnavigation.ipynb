```
from math import sqrt
def calculateStateDerivation(rocket, target):
    currentPosition = rocket[2:]
    currentVelocity = rocket[:2]
    currentVelocityAbs = sqrt(currentVelocity[0] * currentVelocity[0] + currentVelocity[1] * currentVelocity[1])
    currentVelocityDirection = [currentVelocity[0] / currentVelocityAbs, currentVelocity[1] / currentVelocityAbs]

    targetPosition = target[2:]
    targetVelocity  = target[:2]
    targetVelocityEscapeAbs = targetVelocity[0] * currentVelocityDirection[0] + targetVelocity[1] * currentVelocityDirection[1]
    targetVelocityEscape = [targetVelocityEscapeAbs * currentVelocityDirection[0], targetVelocityEscapeAbs * currentVelocityDirection[1]]
    targetVelocityManeuver = [targetVelocity[0] - targetVelocityEscape[0], targetVelocity[1] - targetVelocityEscape[1]]
    targetVelocityManeuverAbs = sqrt(targetVelocityManeuver[0] * targetVelocityManeuver[0] + targetVelocityManeuver[1] * targetVelocityManeuver[1])

    positionR = [targetPosition[0] - currentPosition[0], targetPosition[1] - currentPosition[1]]
    velocityR = [targetVelocity[0] - targetVelocityEscape[0], targetVelocity[1] - targetVelocityEscape[1]]
    positionRAbs = sqrt(positionR[0] * positionR[0] + positionR[1] * positionR[1])#np.linalg.norm(positionR) # abs of vector
    velocityRAbs = sqrt(velocityR[0] * velocityR[0] + velocityR[1] * velocityR[1])#np.linalg.norm(velocityR) # abs of vector

    timeToHit = positionRAbs / velocityRAbs

    omega = targetVelocityManeuverAbs / positionRAbs

    n = 3
    neededAccelerationAbs = n * targetVelocityManeuverAbs / timeToHit
    neededAcceleration = [neededAccelerationAbs * targetVelocityManeuver[0], neededAccelerationAbs * targetVelocityManeuver[1]]
    result = [*neededAcceleration, *currentVelocity]

    return result

def rocketModel(rocket, target):
    currentPosition = np.array(rocket[3:])
    currentVelocity = np.array(rocket[:3])

    targetPosition = np.array(target[3:])
    targetVelocity  = np.array(target[:3])
    currentVelocityAbs = np.linalg.norm(currentVelocity) # abs of vector

    positionR = targetPosition - currentPosition
    positionRAbs = np.linalg.norm(positionR) # abs of vector
    velocityR = targetVelocity - currentVelocity
    velocityRAbs = np.linalg.norm(velocityR) # abs of vector

    omega = 1 / positionRAbs / positionRAbs * np.cross(positionR, velocityR)

    neededAcceleration = -n * velocityRAbs / currentVelocityAbs * \
      np.cross(currentVelocity, omega)
    result = [*neededAcceleration, *currentVelocity]

    if positionRAbs < 1e0: # too close to target, eg. target hit
      #print('target', localData['target']())
      #print('rocket', localData['rocket']())
      #print('state', state)
      #print('currentInput', currentInput)
      localData['rocket']()['finished'] = True
      localData['target']()['finished'] = True
    #'''
    return result

def move(state, statederiv):
  return [state[0] + statederiv[0], state[1] + statederiv[1], state[2] + statederiv[2], state[3] + statederiv[3]]


def AllArrows(targetState, rocketState):
  yield [targetState[2], targetState[3], targetState[0], targetState[1]]
  yield [rocketState[2], rocketState[3], rocketState[0], rocketState[1]]

  stDeriv = calculateStateDerivation(rocketState, targetState)
  yield [rocketState[2] + rocketState[0], rocketState[3] + rocketState[1], *stDeriv[:2]]
  yield [rocketState[2], rocketState[3], targetState[2] + targetState[0] - rocketState[2], targetState[3] + targetState[1] - rocketState[3]]
  yield [rocketState[2], rocketState[3], targetState[2] - rocketState[2], targetState[3] - rocketState[3]]


ax = plotArrows(AllArrows())
ax.set_aspect('equal')
plt.grid(which='both')
plt.show()

for i in range(10):
  rocketState = move(rocketState, calculateStateDerivation(rocketState, targetState))
  targetState = move(targetState, targetStateDerivation)
  ax = plotArrows(AllArrows())
  ax.set_aspect('equal')
  plt.grid(which='both')
  plt.show()
```
