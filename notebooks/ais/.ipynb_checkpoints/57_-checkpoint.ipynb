## Aktivační funkce

$f(x,w,b)=\phi(\sum_i{(\theta_i \cdot x_i)} + b)$

### Tanh

${f(x,w,b) = \tanh(w \cdot i + b)}$

```python
from math import exp, tanh
exp(1)
```

```python
from math import tanh as tgh
def tanh(x, w=1, b=0):
    v = w * x + b
    return tgh(v)
```

Vliv váhy

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4.5, 4.5)
x
w = [1.0, 0.5, 1.5, 2.0]
for wi in w:
    y = list(map(lambda x: tanh(x, wi), x))
    plt.plot(x, y, label=f'w={wi}')
plt.grid()
plt.legend()
```

Vliv bias (posunu)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4.5, 4.5)
x
b = [0, 1.0, 1.5, 2.0]
for bi in b:
    y = list(map(lambda x: tanh(x, b=bi), x))
    plt.plot(x, y, label=f'b={bi}')
plt.grid()
plt.legend()
```

### Sigmoida

$f(x,w,b)=\frac{1}{1+e^{-(w \cdot i + b)}}$

```python
from math import exp
exp(1)
```

```python
def sigmoid(x, w=1, b=0):
    v = w * x + b
    return 1 / (1 + exp(-v))
```

Vliv váhy

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4.5, 4.5)
x
w = [1.0, 0.5, 1.5, 2.0]
for wi in w:
    y = list(map(lambda x: sigmoid(x, wi), x))
    plt.plot(x, y, label=f'w={wi}')
plt.grid()
plt.legend()
```

Vliv bias (posunu)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4.5, 4.5)
x
b = [0, 1.0, 1.5, 2.0]
for bi in b:
    y = list(map(lambda x: sigmoid(x, b=bi), x))
    plt.plot(x, y, label=f'b={bi}')
plt.grid()
plt.legend()
```

### Lineární funkce

```python
def linear(x, w=1, b=0):
    v = w * x + b
    return v
```

### ReLU

```python
def relu(x, w=1, b=0):
    v = w * x + b
    return max(0, v)
```

Vliv váhy

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4.5, 4.5)
x
w = [1.0, 0.5, 1.5, 2.0]
for wi in w:
    y = list(map(lambda x: relu(x, wi), x))
    plt.plot(x, y, label=f'w={wi}')
plt.grid()
plt.legend()
```

Vliv bias (posunu)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4.5, 4.5)
x
b = [0, 1.0, 1.5, 2.0]
for bi in b:
    y = list(map(lambda x: relu(x, b=bi), x))
    plt.plot(x, y, label=f'b={bi}')
plt.grid()
plt.legend()
```

## Optimalizace

https://www.cs.toronto.edu/~hinton/absps/reluICML.pdf

Před rozmachem "hlubokých" neuronových sítí byla aktivační funkce sigmoida velmi populární.
Učení (trénování) neuronové sítě je optimalizačním procesem.
Pro řízení tohoto procesu se používají derivace aktivační funkce (hledání gradientu kriteriální funkce). 

U ReLU aktivační funkce je její derivace v triviálním tvaru

${ \phi'(x) = \begin{cases} 1 & x > 0 \\ 0 & x \leq 0 \end{cases} }$

z výpočetního hlediska se jedná prakticky o tu nejtriviálnější formu, která významně urychluje proces učení.

```python
import graphviz
dot_data = '''
digraph {
    label=""
    ranksep=1
    rankdir="TB"
    layout = dot
    
    subgraph cluster_1 {
        rank = same;
        label="Input Layer"
        fontname="times-bold"
        labeljust = r
        labelangle=180
        color="white"
        Input1 [shape="circle", style="dashed", label="I1"];
        Input2 [shape="circle", style="dashed", label="I2"];
        Bias1 [shape="circle", style="dashed", label="B1"];
        subgraph cluster_1a {
            label="Input Layer"
        }
    }
    subgraph cluster_2 {
        rank = same;
        label="Hidden Layer 1"
        fontname="times-bold"
        labeljust = l
        color="white"
        Neuron11 [shape="circle", label="N2"];
        Neuron12 [shape="circle", label="N1"];
        Bias2 [shape="circle", style="dashed", label="B2"];
    }
    subgraph cluster_3 {
        rank = same;
        label="Hidden Layer 2"
        labeljust = l
        fontname="times-bold"
        color="white"
        Neuron21 [shape="circle", label="N3"];
        Neuron22 [shape="circle", label="N4"];
        Bias3 [shape="circle", style="dashed", label="B3"];
    }
    subgraph cluster_4 {
        rank = same;
        label="Output Layer"
        fontname="times-bold"
        labeljust = l
        color="white"
        Output [shape="circle", style="dashed", label="O "];
    }
    
    
    Bias1 -> Neuron11;
    Bias1 -> Neuron12;
    Bias1 -> Bias2 [color="white"];
    Input1 -> Neuron11;
    Input1 -> Neuron12;
    Input2 -> Neuron11;
    Input2 -> Neuron12;
    
    Bias2 -> Neuron21;
    Bias2 -> Neuron22;
    Bias2 -> Bias3 [color="white"];
    Neuron11 -> Neuron21;
    Neuron11 -> Neuron22;
    Neuron12 -> Neuron21;
    Neuron12 -> Neuron22;
    
    Bias3 -> Output;
    Neuron21 -> Output;
    Neuron22 -> Output;

}
'''
graph = graphviz.Source(dot_data)
graph
graphviz.pipe_string('dot', 'svg', dot_data, encoding='ascii')
```
