```python
import matplotlib.pyplot as plt
#plt.style.use("seaborn-darkgrid")
plt.figure(figsize=(10,5))
plt.plot(df1["W1"], label="W1")
```

```python
def groupW1(data):
    sum = 0
    label = "00"
    for row in data.iterrows():
        index, value = row
        W1 = value["W1"]
        date = value["date1"]
        date = date.replace(" ", ":")

        labels = date.split(":")
        
        if label == labels[1]:
            sum = sum + 5 * W1
        else:
            yield (label, sum / 60)
            sum = 5 * W1
            label = labels[1]
    yield (label, sum / 60)

list(groupW1(df1))
```

```python
rows = list(groupW1(df1))
y = list(map(lambda item: item[1], rows))
plt.step(range(len(y)), y)
```

Příprava dat na NN, zavedení atributu hodina a atribut měsíc

```python
df3 = df.copy()
df3 = df3[df3["date1"].str.startswith("20", na=False)]
df3
```

```python
df3['date1'] = pd.to_datetime(df3['date1'], errors="coerce")
```

```python
df3["date1"].isna().any()
```

```python
df3[df3["date1"].isna()]
```

```python
df3 = df3.dropna(subset = ["date1"])
```

```python
df3['hour'] = df3['date1'].dt.hour
df3['month'] = df3['date1'].dt.month
df3[["date1", "W1", "hour", "month"]]
```

```python
import torch
```

```python
import torch

# Make use of a GPU or MPS (Apple) if one is available.
device = (
    "mps"
    if getattr(torch, "has_mps", False)
    else "cuda"
    if torch.cuda.is_available()
    else "cpu"
)
print(f"Using device: {device}")
```

```python
xval = df3[["month", "hour"]].values
xval
```

```python
x = torch.tensor(xval, dtype=torch.float32)
x[:10]
```

```python
y = torch.tensor(df3["W1"].values, dtype=torch.float32)
y[:10]
```

```python
x.shape[1]
```

```python
import torch.nn as nn

# Create the neural network
model = nn.Sequential(
    nn.Linear(x.shape[1], 200),
    nn.ReLU(),
    nn.Linear(200, 200),
    nn.ReLU(),
    nn.Linear(200, 25),
    nn.ReLU(),
    nn.Linear(25, 1)
)

# PyTorch 2.0 Model Compile (improved performance)
model = torch.compile(model, backend="aot_eager").to(device)

# Define the loss function for regression
loss_fn = nn.MSELoss()

# Define the optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
# Train for 1000 epochs.
for epoch in range(1000):
    optimizer.zero_grad()
    out = model(x).flatten()
    loss = loss_fn(out, y)
    loss.backward()
    optimizer.step()

    # Display status every 100 epochs.
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, loss: {loss.item()}")
```

```python
pred = model(x)
print(f"Shape: {pred.shape}")
print(pred[0:10])
```

```python
import numpy as np
from sklearn import metrics

# Measure RMSE error.  RMSE is common for regression.
score = np.sqrt(metrics.mean_squared_error(pred.cpu().detach(), y.cpu().detach()))
print(f"Final score (RMSE): {score}")
```

```python
import datetime

def draw(day=1, month=11, year=2022):
    start = datetime.datetime(year=year, month=month, day=day)
    end = start + datetime.timedelta(hours=24)
    jj = df3[(df3["date1"] > start) & (df3["date1"] < end)]
    sum_by_hour = jj.groupby('hour')['W1'].sum() / 60 * 5
    plt.plot(sum_by_hour.keys(), sum_by_hour.values)
    #plt.plot(range(len(sum_by_hour)), sum_by_hour)
```
