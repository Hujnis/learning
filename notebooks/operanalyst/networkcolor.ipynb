# Barvení grafu

Hrany v grafu spojují vrcholy, které nesmí sdílet stejnou barvu. Např. při plánování spojujeme události, které nemohou / nesmí probíhat současně, protože využívají stejné zdroje, presonální prostorové, technické vybavení a pod.

Při barvení je cílem dosáhnout minimální počet barev. Minimální počet barev nutných k obarvení grafu je chromatickým číslem grafu.

## Algoritmus

Není znám deterministický algoritmus, který by zaručil tzv. minimální barvení. Jediný algoritmus je vyzkoušení všech variant.

V praxi se používá postupné barvení:
1. určení barvy
2. výběr neobarveného vrcholu a jeho obarvení určenou barvou
3. opakování bodu 2, dokud existují vrcholy, které mohou být obarveny
4. pokud v grafu existují neobarvené vrcholy, návrat do bodu 1

Určení neobraveného vrcholu v bodě 2 se realizuje různými postupy. Jeden z relativně úspěšných je výběr vrcholu, který má nejvíce hran. Je možné taktéž náhodný výběr. Při některých aplikacích se strategie výběru testují při opakovaném řešení úlohy a následně se testuje, který výsledek je lepší. Ten se poté považuje za suboptimální řešení.
