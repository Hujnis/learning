### Prostor $\mathbb{R} \times \mathbb{R}$

Funkce $f: \mathbb{R} \times \mathbb{R} \mapsto \mathbb{R}$, hledáme 
$(x, y): \nabla f(x, y)=(0, 0)$, tedy $\left(\frac{\partial f(x, y)}{\partial x}, 
\frac{\partial f(x, y)}{\partial y}\right)=\left(0,0\right)$

Takové body jsou stacionárními body a mohou (ale nemusí) být extrémy.
Je nutné provést jejich klasifikaci pomocí druhých parciálních derivací. Pro stacionární bod, který je minimem platí $\frac{\partial^2 f(x, y)}{\partial x^2} > 0$ a$\frac{\partial^2 f(x, y)}{\partial y^2} > 0$. Obdobně pro stacionární bod, který je maximem platí $\frac{\partial^2 f(x, y)}{\partial x^2} < 0$ a$\frac{\partial^2 f(x, y)}{\partial y^2} < 0$. Stacionární body, pro které nejsou výše uvedená kritéria splněna jsou sedlovými body.

### Sedlový bod

```python
import matplotlib.pyplot as plt
import numpy as np

def funcA(x, y):
    result = (x * x - y * y)
    return result

plt.rcParams["figure.figsize"] = (12, 6)

x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

X, Y = np.meshgrid(x, y)
Z = funcA(X, Y)

ax = plt.subplot(1, 2, 1, projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='plasma')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax = plt.subplot(1, 2, 2)
CS = ax.contour(X, Y, Z, cmap='plasma')
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('Sedlový bod')

ax.grid(which='both')
```

#### Stacionární body

$$f(x, y)=x^2-y^2$$

$$\frac{\partial f(x, y)}{\partial x} = 2x$$

$$\frac{\partial f(x, y)}{\partial y} = -2y$$

$$2x=0$$

$$-2y=0$$

$$s = (0, 0)$$

#### Klasifikace stacionárního bodu

$$H=\begin{pmatrix}
\frac{\partial^2 f(x, y)}{\partial x^2} & \frac{\partial^2 f(x, y)}{\partial x \partial y} \\
\frac{\partial^2 f(x, y)}{\partial y \partial x} & \frac{\partial^2 f(x, y)}{\partial y^2}
\end{pmatrix}$$

$$\frac{\partial^2 f(x, y)}{\partial x^2} = 2$$

$$\frac{\partial^2 f(x, y)}{\partial x \partial y} = 0$$

$$\frac{\partial^2 f(x, y)}{\partial y \partial x} = 0$$

$$\frac{\partial^2 f(x, y)}{\partial y^2} = -2$$

$$|H|=\left|\begin{matrix}
2 & 0 \\
0 & -2
\end{matrix}\right|=-4$$

#### Závěr

Stacionární bod $(0, 0)$ není extremálním bodem (ani maximum, ani minimum). $(0, 0)$ je sedlovým bodem.

### Metoda Lagraengových multiplikátorů

https://www.youtube.com/watch?v=onsark4J9js

$$\begin{aligned}f(x,y)&=2x-y\\
x^2+y^2&=5\end{aligned}$$

Konstrukce Lagraengovy funkce

$$\begin{aligned}L(x,y,\lambda)&=2x-y+\lambda(x^2+y^2-5)\end{aligned}$$

Parciální derivace

$$\begin{aligned}\frac{\partial L}{\partial x}&=2+\lambda 2x \\
\frac{\partial L}{\partial y}&=-1+\lambda 2y \\
\frac{\partial L}{\partial \lambda}&=x^2+y^2-5\end{aligned}$$

Parciální derivace jsou ve stacionárních bodech nulové

$$\begin{aligned}\frac{\partial L}{\partial x}&=0 \\
\frac{\partial L}{\partial y}&=0 \\
\frac{\partial L}{\partial \lambda}&=0\end{aligned}$$

Odpovídající sada rovnic

$$\begin{aligned}2+\lambda 2x &=0\\
-1+\lambda 2y &=0\\
x^2+y^2-5&=0\end{aligned}$$

Úprava

$$\begin{aligned}1+\lambda x &=0\\
-1+2\lambda y &=0\\
x^2+y^2-5&=0\end{aligned}$$

Příprava na substituci

$$\begin{aligned}x&=-\frac{1}{\lambda}\\
y &=\frac{1}{2\lambda}\\
x^2+y^2-5&=0\end{aligned}$$

Substituce a řešení jedné rovnice

$$\left(-\frac{1}{\lambda}\right)^2+\left(\frac{1}{2\lambda}\right)^2-5=0$$

$$\left(\frac{1}{\lambda^2}\right)+\left(\frac{1}{4\lambda^2}\right)-5=0$$

$$4+1-20\lambda^2=0$$

$$\lambda^2=\frac{1}{4}$$

Kořeny

$$\begin{align}\lambda_1&=-\frac{1}{2}\\
\lambda_2&=\frac{1}{2}\end{align}$$

Odpovídající stacionární body

$$\begin{aligned}x_1&=-\frac{1}{\frac{-1}{2}}\\
y_1 &=\frac{1}{2\frac{-1}{2}}\\
x_2&=-\frac{1}{\frac{1}{2}}\\
y_2 &=\frac{1}{2\frac{1}{2}}\end{aligned}$$

$$\begin{aligned}x_1&=2\\
y_1 &=-1\\
x_2&=-2\\
y_2 &=1\end{aligned}$$

$$A=[-2;1],B=[2;-1]$$

Klasifikace stacionárních bodů pomocí Hessovy matice

$$H=\begin{pmatrix}
\frac{\partial^2 L}{\partial x^2} & \frac{\partial^2 L}{\partial x \partial y} \\
\frac{\partial^2 L}{\partial y \partial x} & \frac{\partial^2 L}{\partial y^2}
\end{pmatrix}$$

$$\begin{aligned}\frac{\partial^2 L}{\partial x^2} &= 2\lambda \\
\frac{\partial^2 L}{\partial x \partial y} &= 0 \\
\frac{\partial^2 L}{\partial y \partial x} &= 0 \\
\frac{\partial^2 L}{\partial y^2} &= 2\lambda \end{aligned}$$

$$|H|=\left|\begin{matrix} 
2\lambda & 0 \\
0 & 2\lambda
\end{matrix}\right|=4\lambda^2$$
